# Trees have Root at the Top and Branches below, inverse of the Traditional Tree

```python
'''
Define the Parent Class for Expressions
'''

class Expressions:
  pass
```

```python
''' 
  Define Sub class and inherit from parent Expressions class
'''  

class Times (Expressions):

  def __init__(self, l, r):
    self.l = l
    self.r = r

  def __str__(self):
    return "(" + str(self.l) + "*" +str(self.r) + ")"

  def eval(self, env):
    return self.l.eval(env) * self.r.eval(env)
```

```python
class Plus (Expressions):

  def __init__(self, l, r):
    self.l = l
    self.r = r

  def __str__(self):
    return "(" + str(self.l) + "+" + str(self.r) + ")"

  def eval(self, env):
    return self.l.eval(env) + self.r.eval(env)
```

```python
class Const (Expressions):

  def __init__(self, val):
    self.val = val

  def __str__(self):
    return str(self.val)

  def eval (self, env):
    return self.val
```

```python
class Variable (Expressions):
    
  def __init__(self,name):
    self.name = name

  def __str__(self):
    return self.name

  def eval(self,env):
    return env[self.name]
```

```python
# 3*(y+x)

e1 = Times(Const(3),Plus(Variable("y"), Variable("x")))
print(e1)
```

```python
# 3*y + x

e2 = Plus(Times(Const(3),Variable("y")),Variable("x"))
print(e2)
```

```python
print(Plus(10,20))
print(Times(10,(10+20)))
print(Variable("x"))
print(Const(10)), Const(10)
```
